```python
#ans 4
"""
The LookupError exception in Python forms the base class for all exceptions that are raised 
when an index or a key is not found for a sequence or dictionary respectively.
We can use LookupError exception class to handle both IndexError and KeyError exception classes.

"""
x=[8,9]
try:
    print(x[5])
except LookupError as e:
    print(e)
```

```python
#ans6
""" best practice for exception handling are as follows:-"""
```

```python
# 1.use always specific exception not standard
try:
    10/0
except ZeroDivisionError as e:
    print(e)
```

```python
# 2.print always a proper message
try:
    10/0
except ZeroDivisionError as e:
    print("i am handling zero division error",e)
```

```python
# 3. always try to log your error
import logging
logging.basicConfig(filename="error.log()",level=logging.ERROR)
try:
    10/0
except ZeroDivisionError as e:
    logging.error("i am handling zero division error {}".format(e))
    
```

```python
#4. always avoid to write multiple exception handling
try:
    10/0
except FileNotFoundError as e:
    logging.error("i am handling filenotfounderror {}".format(e))
except ZeroDivisionError as e:
    logging.error("i am handling zero division error {}".format(e))
```
